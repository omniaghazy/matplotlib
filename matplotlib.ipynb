plt(): has some keywords ⁉

c or color=""

dashes = [length,space]

linewidth = float

antialised = bool      smooth

linestyle or ls ="linestyles "                      change style of line

marker="+"

```python

plt.plot([67,34,26,67,43,2] ,marker = "+" ,dashes =[2,4] , c="k"  )
plt.plot([1,34,56,78,90] , "+" , c="k" , linestyle='dashdot' )




plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("first plot")
plt.show()
```

to change marker color only :

markeredgecolor ="color"

to change marker width(cricle that arounds it) only :

markeredgewidth =float       

```python
plt.plot([1,34,56,78,90] , "s" ,markeredgecolor="c",markeredgewidth=9, c="k" , linestyle='dashdot' )
plt.plot([67,34,26,67,43,2] ,marker = "+" ,markersize =9,dashes =[2,4] , c="k" ,markeredgecolor ="y" )





plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("first plot")
plt.show()
```

```python
plt.plot([67,34,26,67,43,2] ,marker = "o" ,markersize =9,dashes =[2,4] , c="k" ,markeredgecolor ="r" ,markerfacecolor ="y" )





plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("first plot")
plt.show()
```

def xlabel(
  
  xlabel: str,
 fontdict: dict[str, Any] | None=None,
 labelpad: float | None=None,
  *,
  loc: Literal['left', 'center', 'right'] | None=None,
 **kwargs) -> Text

```python
plt.plot([1,2,3,4,5],[12,34,45,67,87] , "b^:")
plt.xlabel(xlabel="x_axis" ,
           fontdict={"size":30 ,"color":"r"} ,
           labelpad= 9   , #[padding ]
           loc="center" #[location]

           )
plt.show()
```

to show your figure:

show()

Display all open figures.

Parameters

block : bool, optional
    Whether to wait for all figures to be closed before returning.

    If True block and run the GUI main loop until all figure windows
are closed.

    If False ensure that all figure windows are displayed and return
immediately.  In this case, you are responsible for ensuring
that the event loop is running to have responsive figures.

    Defaults to True in non-interactive mode and to False in interactive
mode (see .pyplot.isinteractive).

See Also

ion : Enable interactive mode, which shows / updates the figure after
      every plotting command, so that calling show() is not necessary.
ioff : Disable interactive mode.
savefig : Save the figure to an image file instead of showing it on screen

```python
plt.plot([1,2,3,4,5],[12,34,45,67,87] , "b^:")
plt.xlabel(xlabel="x_axis" ,
           size = "x-large" ,  #[fontsize]
           color ="red" ,
           labelpad= 9   , #[padding ]
           loc="center", #[location]
           backgroundcolor ="black" ,
           style ="italic",
           weight = "book",
          #  family ="Chicago"
           )
plt.ylabel(ylabel="y_axis" ,
           size = "x-large" ,  #[fontsize]
           color ="red" ,
           labelpad= 9   , #[padding ]
           loc="center", #[location]
           backgroundcolor ="black" ,
           fontstyle ="italic" ,
           fontweight = 1000    ,# 0===>1000
           fontfamily ="monospace"

           )
plt.show(
    block = False

)
```
